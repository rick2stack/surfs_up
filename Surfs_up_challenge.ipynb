```python
import datetime as dt
import numpy as np
import pandas as pd

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, extract
#from sqlalchemy import extract
#import re

from flask import Flask, jsonify
```

```python
#first create engine for SQL lite
engine=create_engine("sqlite:///hawaii.sqlite")
```

```python
#for this we will just use the automap_base()
#Then we will need to reflect data Base.prepare(engine, reflect=True) to get the columns/tables
Base=automap_base()
Base.prepare(engine, reflect=True)
```

```python
#lets check to make sure which keys we have in the SQL fiel
Base.classes.keys()
```

```python
#Create a Measurement filter
Measurement = Base.classes.measurement
#Create a station filter
Station = Base.classes.station
#create a session, this is what you can do querys one
session = Session(engine)
```

```python
#Part 1, 
#3---- In Step 1, write a query that filters the date column from the Measurement table to retrieve all the temperatures for the month of June.
#order_by(Measurement.date.desc()) if you need to sort by desc
#can't filter out with regular expression, but extract function works 
# https://stackoverflow.com/questions/11616260/how-to-get-all-objects-with-a-date-that-fall-in-a-specific-month-sqlalchemy/31641488
june_measurement=session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date)==6).all()
june_measurement
```

```python
#4-----In Step 2, convert the June temperatures to a list.
june_temps_list=[temps for date, temps in june_measurement]
june_temps_list
```

```python
#5-----In Step 3, create a DataFrame from the list of temperatures for the month of June.
columns= ["June Temp"]
june_temps_df=pd.DataFrame(june_temps_list, columns=columns)
june_temps_df
```

```python
#In Step 4, generate the summary statistics for the June temperatures DataFrame
june_temps_df.describe()
```

```python
# Part 2, 
#2---In Step 6, write a query that filters the date column from the Measurement table to retrieve all the temperatures for the month of December.
#can't filter out with regular expression, but extract function works 
# https://stackoverflow.com/questions/11616260/how-to-get-all-objects-with-a-date-that-fall-in-a-specific-month-sqlalchemy/31641488
december_measurement=session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date)==12).all()
#3---In Step 7, convert the December temperatures to a list.
december_temps_list=[temps for date, temps in december_measurement]
december_temps_list
```

```python
#Part 2
#4----In Step 8, create a DataFrame from the list of temperatures for the month of December.
columns= ["December Temp"]
december_temps_df=pd.DataFrame(december_temps_list, columns=columns)
december_temps_df
```

```python
#Part 2
#5--- In Step 9, generate the summary statistics for the December temperatures DataFrame.
december_temps_df.describe()
```

```python
##Additional Queries for Analysiss
#precipitation query
december_prcp=session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date)==12).all()
june_prcp=session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date)==6).all()
#precipitation List
december_prcp_list=[prcp for date, prcp in december_prcp]
june_prcp_list=[prcp for date, prcp in june_prcp]
```

```python
columns= ["June precipitation"]
June_precipitation_df=pd.DataFrame(june_prcp_list, columns=columns)
June_precipitation_df.describe()
```

```python
columns= ["December precipitation"]
december_precipitation_df=pd.DataFrame(december_prcp_list, columns=columns)
december_precipitation_df.describe()
```
